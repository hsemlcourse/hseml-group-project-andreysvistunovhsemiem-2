# 03 — Эксперименты

Сравнение 5 моделей на едином стратифицированном сплите (seed из `.env`).

```python
import matplotlib.pyplot as plt
import pandas as pd

from src.config import settings
from src.modeling import run_experiments, train_and_eval_baseline
```

```python
baseline = train_and_eval_baseline()
experiments = run_experiments()
reports = {**baseline, **experiments}
table = pd.DataFrame({k: v.to_dict() for k, v in reports.items()}).T
table = table.sort_values('roc_auc', ascending=False)
table
```

```python
settings.images_dir.mkdir(parents=True, exist_ok=True)
fig, ax = plt.subplots(figsize=(7, 3.5))
table['roc_auc'].plot(kind='barh', ax=ax, color='#4a7ab5')
ax.set_xlabel('ROC-AUC')
ax.set_title(f'ROC-AUC по моделям (test, seed={settings.random_state})')
ax.invert_yaxis()
fig.tight_layout(); fig.savefig(settings.images_dir / 'experiments_roc_auc.png', dpi=120); plt.show()
```

## Выводы

- Baseline LogReg без FE: ROC-AUC ≈ 0.763.
- FE даёт линейной модели +0.005 ROC-AUC (`logreg_fe`).
- KNN (k=25) сопоставим с LogReg по ROC-AUC, но сильно теряет в recall.
- Ансамбли деревьев — сильнее линейных: RF ≈ 0.764, LightGBM ≈ 0.766, XGBoost ≈ 0.774 (лучший результат).
- Выбор финальной модели + тюнинг — в CP2.
